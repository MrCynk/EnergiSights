### Séparation du dataset

```python
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
```

```python
y = np.log1p(df["TotalGHGEmissions"]) #df["SiteEnergyUse(kBtu)"]
X = df.drop(["TotalGHGEmissions"], axis = 1)
```

```python
y.isnull().sum()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

### Imputations et Pipeline
### Features selection

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.metrics import mean_absolute_error, r2_score, mean_squared_error
```

```python
from xgboost import XGBRegressor
```

```python
X.columns
```

```python
num_var = ["NumberofFloors", "PropertyGFATotal",
           "ENERGYSTARScore", "built_Age", "building_Ratio"]

cat_var = ["BuildingType", "PrimaryPropertyType", "Neighborhood"]
```

```python
num_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy='median'))
])
```

```python
cat_pipeline = Pipeline(steps=[
    ("imputer", SimpleImputer(strategy='most_frequent')),
    ("onehot", OneHotEncoder(handle_unknown="ignore"))
])
```

```python
preprocess = ColumnTransformer(
    transformers = [
        ("num", num_pipeline, num_var),
        ("cat", cat_pipeline, cat_var)
    ]
)
```

```python
y.shape
```

```python
model_xgb = Pipeline(steps=[("preprocess", preprocess),
                             ("xgb_model", XGBRegressor(
                                 n_estimators=800,
                                 learning_rate=0.03,
                                 max_depth=2,
                                 subsample=0.8,
                                 colsample_bytree=0.8,
                                 reg_alpha=0.1,
                                 reg_lambda=3.0,
                                 #min_samples_split=10,
                                 #min_samples_leaf=5,
                                 #max_features="sqrt",
                                 random_state=42,
                                 n_jobs=-1
                             ))
])
```

```python
model_xgb.fit(X_train, y_train)
```

```python
y_pred = model_xgb.predict(X_test)
```

```python
y_pred
```

```python
model_xgb.score(X_train, y_train)
```

```python
model_xgb.score(X_test, y_test)
```

```python
print("R2:", r2_score(y_test, y_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_pred)))
print("MAE:", np.sqrt(mean_absolute_error(y_test, y_pred)))
```

```python
scores = cross_val_score(model_xgb, X, y, cv=5, scoring="r2")
print(scores)
```

```python
print("Mean CV:",scores.mean())
print("Ecart-type:",scores.std())
```
